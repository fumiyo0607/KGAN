```python
import pandas as pd
from collections import defaultdict
from tqdm import tqdm
import os
import numpy as np
```

```python
data_path = './data/last-fm/'
row_data_path = './row_data/LFM-1b/'
```

#### train.txt の読み込み

```python
train_data = defaultdict(list)
f = open( data_path + 'train.txt')
line = f.readline()

while line :

    data = line.strip()
    data_list = data.split()

    user = data_list[0]
    items = data_list[1:]
    train_data[user] = items
    
    line = f.readline()
    
f.close()
```

#### test.txt の読み込み

```python
test_data = defaultdict(list)
f = open(data_path + 'test.txt')
line = f.readline()

user_count = 0

while line :

    data = line.strip()
    data_list = data.split()

    user = data_list[0]
    items = data_list[1:]
    train_data[user] = items
    
    line = f.readline()
    
f.close()
```

#### train data とtest data を結合する

```python
all_data = train_data


for user, items in test_data.items():
    all_data[user] = all_data[user] + items
    
```

```python
all_data = train_data

for user, items in test_data.items():
    all_data[user] = all_data[user] + items
    
```

```python
user_list = []
item_list = []

for user, items in all_data.items():
    
    for item in items:
        user_list.append(user)
        item_list.append(item)
```

```python
data_df = pd.DataFrame(
    data = {
        'user_id' : user_list,
        'item_id' : item_list
    },
    columns=['user_id', 'item_id']
)
```

```python
data_df.head()
```

#### item_list.txt の読み込み

```python
item_df = pd.read_csv( data_path + 'item_list.txt', sep=' ')
item_df.head(5)
```

#### item_id と org_id (last-fm) を結合

```python
data_df['item_id'] = data_df['item_id'].astype(np.int64)
```

```python
df = pd.merge(
                    data_df, item_df, 
                    left_on='item_id', 
                    right_on='remap_id', 
                    how='left'
             ).drop(columns='remap_id')
```

```python
df.head(10)
```

#### last fm data の読み込み

```python
last_fm_stack_df = pd.read_csv(
    row_data_path + 'LFM-1b_tracks.txt', 
    sep='\t', 
    header=None, 
    names=('org_id', 'track_name', 'artist_id')
)
```

```python
last_fm_stack_df.head(5)
```

```python
last_fm_artists_df = pd.read_csv(
    row_data_path + 'LFM-1b_artists.txt', 
    sep='\t',
    header=None, 
    names=('artist_id', 'artist_name')
)
```

```python
last_fm_artists_df.head(5)
```

```python
last_fm_df = pd.merge(
    last_fm_stack_df, last_fm_artists_df,
    on='artist_id',
    how='left'
)
```

```python
last_fm_df.head(5)
```

#### モデルの学習用データとLast-FMのデータを結合

```python
all_data_df = pd.merge(
    df, last_fm_df,
    on='org_id',
    how='left'
)
```

```python
all_data_df.head(10)
```
